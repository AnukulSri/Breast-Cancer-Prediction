## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df['diagnosis'].value_counts()
```

```python
sns.heatmap(df.corr())
```

## Model Training

```python
print(df.columns.tolist())
```

```python
x= df[['radius_mean', 
       'texture_mean', 'perimeter_mean', 
       'area_mean', 'smoothness_mean', 'compactness_mean', 
       'concavity_mean', 'concave points_mean', 'symmetry_mean', 
       'fractal_dimension_mean', 'radius_se', 'texture_se', 
       'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 
       'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 
       'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 
       'smoothness_worst', 'compactness_worst', 'concavity_worst', 
       'concave points_worst', 'symmetry_worst', 
       'fractal_dimension_worst']].values
y = df[['diagnosis']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,Y_train)
```

```python
dtc_pred = dtc.predict(X_test)
dtc_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Confusion matrix :\n",confusion_matrix(Y_test,dtc_pred))
```

```python
print(classification_report(Y_test,dtc_pred))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
lr_pred
```

```python
print("Confusion matrix of logistic regression: \n",confusion_matrix(Y_test,lr_pred))
```

```python
print("Classification report of logistic regression: \n",classification_report(Y_test,lr_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
```

```python
rfc_pred = rfc.predict(X_test)
rfc_pred
```

```python
print("Confusion matrix of random forest classifier: \n",confusion_matrix(Y_test,rfc_pred))
```

```python
print("Classification report of random forest classifier :\n",classification_report(Y_test,rfc_pred))
```
